````python
# import library yang dibutuhkan
import PyPDF2
import glob
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
# init list
docs = []
````

````python
# init variable
total_num_page = []
total_file_name = []

# perulangan pada file berekstensi pdf untuk menghitung halaman
for fileName in glob.glob(os.path.join("*.pdf")):
  total_file_name.append(fileName)
  #  getcwd() returns current working directory of a process, rb means read binary
  with open(os.path.join(os.getcwd(), fileName), "rb") as pdfObject:
    # init pdfReader from PdfFileReader of pdfObject
    pdfReader = PyPDF2.PdfReader(pdfObject)
    # nomor halaman
    numPage = len(pdfReader.pages)
    total_num_page.append(numPage)
    # menampilkan nama file berikut halaman
    print(fileName, "\t", numPage, " Halaman")
    # perulangan untuk ekstark text dari object
    for page in range(numPage):
      # read per page and append with extract_text
      pageObject = pdfReader.pages[page]
      docs.append(pageObject.extract_text())
````

````python
# menampilkan banyak dataset
print("Total dataset : ", len(fileName))
````

````python
# menampilkan isi docs index ke-0
print(docs[0])
````

````python
# init list untuk cleaning dataset
docs_clean = []
````

````python
# perulangan pada tiap line(dalam hal ini kalimat) dari list docs
for sentence in docs:
    # remove unicode
    docs_tmp = re.sub(r'[^\x00-\x7F]+', ' ', sentence)
    # remove mentions
    docs_tmp = re.sub(r'@\w+', '', docs_tmp)
    # lowercase the document
    docs_tmp = docs_tmp.lower()
    # remove punctuations
    docs_tmp = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', docs_tmp)
    # Lowercase the numbers
    docs_tmp = re.sub(r'[0-9]', '', docs_tmp)
    # remove the doubled space
    docs_tmp = re.sub(r'\s{2,}', ' ', docs_tmp)
    # append
    docs_clean.append(docs_tmp)
````

````python
# menampilkan hasil cleaning dari list docs_clean index ke-0
print(docs_clean[0])
````

````python
# kita perlu melakukakn tokenizing yang berfungsi :
# pembentukan term dari sederetan karakter. 
# term adalah karakter alfanumerik dengan batasan panjang karakter tertentu. 
# dibatasi dengan spasi atau karakter tertentu

# init list untuk tokenizing
docs_tokenize = []
````

````python
# download package punkt dari library nltk
nltk.download('punkt')
````

````python
# proses tokenizing
docs_tokenize = [word_tokenize(words) for words in docs_clean]
````

````python
# menampilkan hasil tokenizing dari list docs_tokenize index ke-0
docs_tokenize[0]
````

````python
# init object StemmerFactory() dan pembuatan variable stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
````

````python
# init final list
docs_final = []
````

````python
# show before stemming, 10 data
for words in docs_tokenize[:10]:
  print(words)
````

````python
docs_tokenize[:10]
````

````python
# stemming process
for words in docs_tokenize:
  docs_stemming = stemmer.stem(str(words))
  docs_final.append(docs_stemming)
````

````python
def SearchDocument(query):
    q =[query]
    # embed the query for calcluating the similarity
    Q_Train =embed(q)
    
    #imported_m = tf.saved_model.load('/home/zettadevs/GoogleUSEModel/TrainModel')
    #loadedmodel =imported_m.v.numpy()
    # Calculate the Similarity
    linear_similarities = linear_kernel(Q_Train).flatten() 
    #Sort top 10 index with similarity score
    Top_index_doc = linear_similarities.argsort()[:-11:-1]
    # sort by similarity score
    linear_similarities.sort()
    a = pd.DataFrame()
    for i,index in enumerate(Top_index_doc):
        a.loc[i,'index'] = str(index)
        a.loc[i,'result'] = docs_final[index] ## Read File name with index from File_data DF
    for j,simScore in enumerate(linear_similarities[:-11:-1]):
        a.loc[j,'score'] = simScore
    return a
````
